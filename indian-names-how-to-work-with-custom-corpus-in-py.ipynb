**Importing a word list file**

#loading libraries
import nltk.data
  
nltk.data.load('corpora/Indian/first_name.txt', format ='raw')

```python
# opening the file in read mode
my_file = open("/kaggle/input/indian-names-corpus-nltk-data/Indian Names/first_name.txt", "r")
# reading the file
data = my_file.read()
  
# replacing end of line('/n') with ' ' and
# splitting the text it further when '.' is seen.
data_into_list = data.split("\n")
  
# printing the data
print(data_into_list)
my_file.close()
```
