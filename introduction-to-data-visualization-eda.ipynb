## Part A) Create a dataset with 12 features and 3 classes using make_classification. Name the features F1, F2, ... upto F12. F1: Can hold two values (True/False) F2: Can hold two values (Type 1 or Type 2) F3: Can hold three values (A or B or C) F4: Can hold three values (HIGH / MEDIUM / LOW) Features F5 to F12 will have numeric values generated during the creation process.


```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=1000,n_features=12,n_redundant=0,n_informative=5,n_classes=3,random_state=1,n_clusters_per_class=1)
import pandas as pd
data1 = pd.DataFrame(X)

# Dataframe done

# Change column name
data1 = data1.rename(columns={0:'F1',1:'F2',2:'F3',3:'F4',4:'F5',5:'F6',6:'F7',7:'F8',8:'F9',9:'F10',10:'F11',11:'F12'})

# Change the feature column value
# F1
data1.loc[data1.F1 >= 0,'F1'] = True
data1.loc[data1.F1 < 0,'F1'] = False

# F2
conditions = [(data1['F2'] >= 0),
              (data1['F2'] < 0)]
values = ['Type 1','Type 2']
data1['F2'] = np.select(conditions,values)
data1

# F3
PE_Conditions = [
    (data1['F3'] < -1),
    (data1['F3'] >= -1) & (data1['F3'] <=1),
    (data1['F3'] > 1)
]
values = ['C','B','A']
data1['F3'] = np.select(PE_Conditions,values)
data1

# F4
PE_Conditions = [
    (data1['F4'] < -1),
    (data1['F4'] >= -1) & (data1['F4'] <=1),
    (data1['F4'] > 1)
]
values = ['Low','Medium','High']
data1['F4'] = np.select(PE_Conditions,values)
data1
```

## Part B) Use either of the above datasets to show the usage of distplot, jointplot, pairplot, rugplot, catplot, barplot, countplot, violinplot, striplot, swarmplot and facetgrid plots.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *
```

```python
sns.displot(data1)
plt.show()
```

A Distplot or distribution plot, depicts the variation in the data distribution. Seaborn Distplot represents the overall distribution of continuous data variables. In this plot, the entire distribution of our dataset is shown.

```python
sns.jointplot(data=data1, x="F12", y="F5",color='red')
```

A Jointplot comprises three plots. Out of the three, one plot displays a bivariate graph which shows how the dependent variable(Y) varies with the independent variable(X). Another plot is placed horizontally at the top of the bivariate graph and it shows the distribution of the independent variable(X). In our plot, we have plotted the F12 feature on the x-axis and F5 on the y-axis. The complete plot shows how one varies in relation to the other.

```python
sns.pairplot(data1,hue='F1')
```

Pairplot shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots. In our plot, we have considered the F1 feature to show its relationship between F5 till F12.


```python
sns.scatterplot(data=data1, x="F10", y="F11")
sns.rugplot(data=data1, x="F10", y="F11",hue="F3",height=.1)
```

A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualise the distribution of the data. As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot. Since our feature 3 had 3 parameters, we have added it as hue and the main plot is contructed for F10 and F11. It shows the distribution of one in relation to other.

```python
g = sns.catplot(x="F4", y="F5", data=data1,kind='boxen',hue='F3')
```

Catplot provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. In our plot, we have shown the relation between F4 and F5.


```python
sns.barplot(x="F1", y="F12", data=data1,hue='F4')
```

A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. In our plot, we have shown the categories of the F1 column- which has two values: True and False.

```python
f, ax = plt.subplots(2,2)
sns.countplot(x='F1', data=data1, ax=ax[0,0])
sns.countplot(x='F2', data=data1, ax=ax[0,1])
sns.countplot(x='F3', data=data1, ax=ax[1,0])
sns.countplot(x='F4', data=data1, ax=ax[1,1])
f.tight_layout(pad=1.0)
```

A barplot shows the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. In our plot, we have created the countplot for the first 4 features of our dataset.

```python
sns.violinplot(x="F2", y="F8", hue="F3",data=data1, palette="muted")
```

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable. In our plot, we have plotted F2(categorical) against F8(numerical) and added the hue as F3.

```python
sns.stripplot(x="F2", y="F5", hue="F1",data=data1, palette="Set2",dodge=True)
```

A strip plot is a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values. The values are plotted as dots along one unique axis, and the dots with the same value can overlap. We have plotted the F2 in the form of a strip plot here.

```python
ax = sns.swarmplot(x="F4", y="F7", data=data1,edgecolor='gray',size=3.4)
```

A swarmplot shows all the data points and that helps to understand the distribution in a better manner. It also helps to understand how the data is distributed across a categorical attribute and how the continuous variable is varying within a category. In our plot, we are trying to understand the distribution of the F4 feature better using the swarmplot.

```python
g = sns.FacetGrid(data1, col="F3")
g.map(sns.histplot, "F8")
```

FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels. In out plot, we have visualized the F3 feature having 3 values- A, B and C- and their distribution visualized by using a histogram.

## Part C) Load the Iris Dataset, and explore - scatterplot, scatter_3d, heatmap, boxplot, kdeplot etc.


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
iris
```

```python
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
plt.figure(2, figsize=(8, 6))
plt.clf()
```

```python
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

A scatterplot shows the relationship between two quantitative variables measured for the same individuals. The values of one variable appear on the horizontal axis, and the values of the other variable appear on the vertical axis. Each individual in the data appears as a point on the graph. In our plot, we have depicted the relationship between the sepal length and the sepal width of the iris data set.

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
plt.show()
```

scatter_3d plots individual data in three-dimensional space. Data points that tend to rise together suggest a positive correlation. Data points that tend to rise as other data points tend to decline suggests a negative correlation. Outliers fall far from the main group of data points. In our plot, after transforming the data set, we have plotted a scatter plot for the first three features. 

```python
import pandas as pd
import seaborn as sns
iris=pd.read_csv("../input/iris/Iris.csv")
iris
```

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

A box and whisker plot—also called a box plot—displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In our plot, we have plotted the boxplot for all three varieties of the iris data set against their petal length. 

```python
plt.figure(figsize=(7,4))
sns.heatmap(iris.corr(),annot=True,cmap='summer')
```

A heatmap is a graphical representation of data that uses a system of color-coding to represent different values. Heatmaps are used in various forms of analytics but are most commonly used to show user behavior on specific webpages or webpage templates. In our plot, we have shown the graphical representation of the different columns of the iris dataset. Varying colors in the heatmap depict the differences in various features corresponding to different classes, for example, the sepal length for multiple categories. It also shows that sometimes sepal lengths tend to be much bigger than petal lengths.

```python
iris = sns.load_dataset("iris")
sns.kdeplot(data=iris, x="sepal_length")
# Setting the X and Y Label
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```

Kdeplot is a Kernel Distribution Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables i.e. we can plot for the univariate or multiple variables altogether. Using the Python Seaborn module, we can build the Kdeplot with various functionality added to it. In our plot, we have plotted the kde of the sepal length of the iris data set. It shows that at the median value of the sepal length, the probability density is the highest.
